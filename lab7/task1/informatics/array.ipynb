D. Количество элементов, больших предыдущего

```python
n=int(input())
arr=list(map(int, input().split()))
x=0
for i in range(1, n):
    if arr[i] > arr[i - 1]:
        x+=1
print(x)
```

E. Есть ли два элемента с одинаковыми знаками

```python
n=int(input())
arr=list(map(int, input().split()))
for i in range(1, n):
    if (arr[i]<0 and arr[i-1]<0) or (arr[i]>0 and arr[i-1]>0):
        print("YES")
        break
else:
    print("NO")
```

F. Количество элементов больших обоих соседей

```python
n=int(input())
arr=list(map(int, input().split()))
x=0
for i in range(2, n):
    if arr[i-2]<arr[i-1] and arr[i]<arr[i-1]:
        x+=1
print(x)
```
