A. Минимум 4 чисел

```python
def min(arr):
    min=arr[0]
    for i in arr:
        if i<min:
            min=i
    print(min)

arr=list(map(int, input().split()))
min(arr)
```

B. Степень

```python
def power(a,b):
    print(a**b)
a, b = map(int, input().split())
power(a,b)
```

C. Исключающее ИЛИ

```python
def xor(a,b):
    if(a==1 or b==1)or(a==1 and b==1):
        print(1)
    else:
        print(0)
a, b = map(int, input().split())
xor(a,b)
```
