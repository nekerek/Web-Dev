A. Четные числа 

```python
a=int(input())
b=int(input())
for i in range(a, b+1):
    if i%2==0:
        print(i, end=" ")
```

B. Остаток

```python
a, b, c, d = map(int, input().split())

e=""
for i in range(a, b+1):
    if i%d==c:
        e+=str(i)+' '
print(e)
```

C. Квадраты

```python
a=int(input())
b=int(input())
c = ""
for i in range(a, b+1):
    for j in range(1, b+1):
        if j**2 == i:
            c += str(i)+' '
print(c)
```

D.  Цифра в числе

```python
a=str(input())
b=str(input())
x=0
for i in a:
    if i==b:
        x+=1
print(x)
```

E. Сумма цифр

```python
n = str(input())
sum = 0
for i in n:
    sum+=int(i)
print(sum)
```

F.  Переверни число

```python
n=str(input())
print(n[::-1])
```

G. Минимальный делитель

```python
a=int(input())
for i in range(2, a+1):
    if a%i==0:
        print(i)
        break
```

H. Делители числа

```python
a=int(input())
c=""
for i in range(1, a+1):
    if a%i==0:
        c+=str(i)+' '
print(c)
```

I.  Количество делителей

```python
a=int(input())
x=0
for i in range(1, a+1):
    if a%i==0:
        x+=1
print(x)
```

J. Сумма ста

```python
sum=0
for i in range(0,101):
    a=int(input())
    sum+=a
print(sum)
```

K. Сумма чисел

```python
n=int(input())
sum=0
for i in range(n):
    a=int(input())
    sum+=a
print(sum)
```
