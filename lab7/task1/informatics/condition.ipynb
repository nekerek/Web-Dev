B. Високосный год

```python
a=int(input())
if(a%4==0 and a%100!=0) or a%400==0:
    print("YES")
else:
    print("NO")
```

C. Тестирующая система

```python
a=int(input())
b=int(input())
if (a==1 and b==1)or(a!=1 and b!=1):
    print("YES")
else:
    print("NO")
```

D. Знак числа

```python
x=int(input())
if x>0:
    print(1)
elif x<0:
    print(-1)
else:
    print(0)
```
