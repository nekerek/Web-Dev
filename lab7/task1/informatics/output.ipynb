A. Гипотенуза


```python
from math import sqrt, pow
a = int(input())
b = int(input())

print(sqrt(pow(a, 2) + pow(b, 2)))
```

B. Следующее и предыдущее

```python
a=int(input())
b=a+1
c=a-1
print("The next number for the number", a, "is", b)
print("The previous number for the number", a, "is", c)
```
