A. Список квадратов

```python
n=int(input())
x=1
while x*x <= n:
    print(x*x)
    x+=1
```

B. Минимальный делитель

```python
n=int(input())
x=2
while x<=n:
    if n%x==0:
        print(x)
        break
    x+=1
```

C. Список степеней двойки

```python
n=int(input())
x=0
while 2**x<=n:
    print(2**x)
    x+=1
```

D. Точная степень двойки

```python
n=int(input())
x=0
while 2**x<=n:
    if 2**x==n:
        print("YES")
        break
    x+=1
else:
    print("NO")
```

E. Двоичный логарифм

```python
n=int(input())
k=0
x=1
while x<n:
    x*=2
    k+=1
print(k)
```
